```python
import tensorflow as tf
import keras
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model1 = keras.models.load_model("model1.hdf5")
model2 = keras.models.load_model("model2.hdf5")
samples = np.load("samples.npy")
```

```python
#MNIST Daten laden
from keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Dimensionen anpassen
x_train_reshaped = x_train.reshape((60000, 28*28))
x_train_reshaped[0].shape
```

```python
samples_reshaped = samples[:,:,:,0]
samples_reshaped = samples.reshape((100, 28*28))
samples_reshaped = samples_reshaped*255
samples_reshaped.shape
```

```python
#Indices speichern
samples_indices = []

for x in samples_reshaped:
    i = 0
    for y in x_train_reshaped:
        if np.array_equal(x,y):
            samples_indices.append(i)
        i = i + 1
```

```python
labels = y_train[samples_indices]
```

```python
labels
```

```python
list1= model1.predict(samples)
list2= model2.predict(samples)
```

```python
liste1 = np.argmax(list1 , axis=1)
liste2 = np.argmax(list2 , axis=1)
```

```python
print(liste1)
print(liste2)
print(labels)
```

```python
qualität = list()
for x in range(0, len(labels)):
    if labels[x] == liste1[x] and labels[x] != liste2[x]:
        qualität.append ("Model 1 Richtig")
        
    if labels[x] == liste2[x] and labels[x] != liste1[x]:
        qualität.append("Model 2 Richtig")
        
    if labels[x] == liste1[x] and labels[x] == liste2[x]:
        qualität.append("Model 1 & 2 Richtig")
        
    if labels[x] != liste1[x] and labels[x] != liste2[x]:
        qualität.append("Model 1 & 2 Falsch")
```

```python
qualität
```

```python
len(qualität)
```

```python
d = {'Model1': liste1, 'Model2': liste2,'Label': labels, 'Resultat':qualität}
df = pd.DataFrame(data=d)
df
```

```python
df.sort_values(by=['Label'])
```
